### Kind of a quickstart for nemo ðŸ¤¨

```python
from nemoguardrails import LLMRails, RailsConfig
from dotenv import load_dotenv
from rich import print
load_dotenv("../secrets/local.env")
```

```python
# load the config
config = RailsConfig.from_path("../guardrails_config")
config
```

```python
rails = LLMRails(config)

response = rails.generate(messages=[
    {
    "role": "user",
    "content": "What are the benefits of investing in a basket?"
    }
])

print(response["content"])
```

```python
# Explain the rails
info = rails.explain()
print(info)
```

```python
print("**Colang History**")
print(info.colang_history)
print("**LLM Calls**")
print(info.print_llm_calls_summary())
print("**Prompt**")
print(info.llm_calls[1].prompt)
```
